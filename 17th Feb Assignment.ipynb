```python
# 1.Ans.

MongoDB is a popular open-source, document-based NoSQL database system that is 
designed to store, manage and access unstructured and semi-structured data, such as 
JSON documents, binary data, text, and other data types. MongoDB uses a flexible document
data model that allows for dynamic schemas, making it well-suited for applications with 
evolving data structures.

Non-relational databases, also known as NoSQL databases, are database management 
systems that are designed to handle unstructured or semi-structured data that 
doesn't fit neatly into the rows and columns of traditional relational databases. 
NoSQL databases use a variety of data models, such as document-based, key-value, graph-based, 
and column-family, to store and manage data.

There are a few scenarios where MongoDB may be preferred over SQL databases:

Flexibility: MongoDB provides a flexible schema that can handle dynamic and 
rapidly changing data. This makes it ideal for applications that require frequent 
schema changes or for projects where the data structure is not fully defined at the outset.

Scalability: MongoDB is highly scalable, both horizontally and vertically. 
It can handle large volumes of data and can scale out easily by adding more nodes to a cluster.

Performance: MongoDB is designed for high-performance, with features such as in-memory 
computing and a powerful query language that can handle complex queries quickly.

Cloud-native: MongoDB is well-suited for cloud-native applications, with features such as
automatic sharding and data distribution, automatic failover, and easy integration with 
cloud platforms like AWS and Azure.
```

```python
# 4.Ans.

import pymongo
client = pymongo.MongoClient('mongodb://localhost:12345/')

db = client['mydatabase']
collection = db['mycollection']

document = {'name': 'Sk', 'age': 21}
result = collection.insert_one(document)

found_doc = collection.find_one({'_id': result.inserted_id})
print(found_doc)

documents = [
    {'name': 'Ks', 'age': 30},
    {'name': 'Sm', 'age': 40},
    {'name': 'Pm', 'age': 50}
]
result = collection.insert_many(documents)

found_docs = collection.find({'age': {'$gte': 30}})
for doc in found_docs:
    print(doc)
```

```python
# 5.Ans.

In MongoDB, the find() method is used to query the database and retrieve documents
that match a specific criteria. The method takes one or two arguments.

A query object that specifies the selection criteria for the documents to be retrieved.
This is a dictionary that contains one or more key-value pairs that represent the fields
and values to be matched.

An optional projection object that specifies the fields to be included or excluded from 
the returned documents. This is a dictionary that contains one or more key-value pairs that
represent the fields and their values, where 1 indicates to include the field, and 0 indicates 
to exclude the field.

  Code :-
    
    import pymongo
    
client = pymongo.MongoClient('mongodb://localhost:12345/')

db = client['mydatabase']
collection = db['mycollection']

collection.insert_many([
    {'name': 'Sk', 'age': 25},
    {'name': 'Ks', 'age': 30},
    {'name': 'Sm', 'age': 40},
    {'name': 'Pm', 'age': 50}
])

all_documents = collection.find()
print('All documents:')
for doc in all_documents:
    print(doc)



    
    
```

```python
# 6.Ans.

In MongoDB, the sort() method is used to sort the results of a query in ascending

or descending order based on one or more fields. The method takes one or more arguments:

A sort key that specifies the field or fields to sort by. This can be a string that 
represents a field name, or a list of tuples that represent field names and their sort 
orders (1 for ascending and -1 for descending).

An optional direction argument that specifies the sort direction. This can be 1 for ascending 
order (the default) or -1 for descending order.

      Example:
        
import pymongo
    
client = pymongo.MongoClient('mongodb://localhost:12345/')

db = client['mydatabase']
collection = db['mycollection']

collection.insert_many([
    {'name': 'Sk', 'age': 25},
    {'name': 'Ks', 'age': 30},
    {'name': 'Sm', 'age': 40},
    {'name': 'Pm', 'age': 50}
])

name_sort = [('name', 1)]
name_documents = collection.find().sort(name_sort)
print('Documents sorted by name in ascending order:')
for doc in name_documents:
    print(doc)
```
